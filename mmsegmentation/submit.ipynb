```python
import json
import numpy as np

class NumpyEncoder(json.JSONEncoder):
    """ Custom encoder for numpy data types """
    def default(self, obj):
        if isinstance(obj, (np.int_, np.intc, np.intp, np.int8,
                            np.int16, np.int32, np.int64, np.uint8,
                            np.uint16, np.uint32, np.uint64)):
            return int(obj)
        elif isinstance(obj, (np.float_, np.float16, np.float32, 
                              np.float64)):
            return float(obj)
        elif isinstance(obj, (np.ndarray,)): 
            return obj.tolist()
        return json.JSONEncoder.default(self, obj)

def write_to_jsonl(data_list, filename):
    with open(filename, 'w') as file:
        for data_dict in data_list:
            json_str = json.dumps(data_dict, cls=NumpyEncoder)
            file.write(json_str + '\n')


import cv2
import numpy as np
import torch

def mask_to_polygon(mask_np):
    # Find contours in the binary mask
    
    mask_np = mask_np.astype(np.uint8)
    
    contours, _ = cv2.findContours(mask_np, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # Convert contours to polygons
    polygons = []
    for contour in contours:
        polygon = [tuple(point[0]) for point in contour]
        polygons.append(polygon)
    
    return polygons
```

```python
jsons = []

idxs = []

for output, id in tqdm(zip(results, img_paths)):
    output = output.copy()
    imageName = id['filename']
    
    height = output.shape[0]
    width = output.shape[1]
    
    json_dat = {}
    shapes = []
    msks = []
    #output = cv2.resize(output, (width, height))
    
    for i in range(19):
        label = idx_to_class[i]
        msk = (output==i).copy()
        msk[msk>=0.5] = 1
        msk[msk<0.5] = 0
        
        polygons = mask_to_polygon(msk)
        if polygons:
            for poly in polygons:
                dat = {"label": label, "points": [list(p) for p in poly], "shape_type": "polygon"}
                shapes.append(dat)
        
        msks.append(msk)
        
        #break
    
    json_dat['imageName'] = imageName
    json_dat["imageHeight"] = height
    json_dat["imageWidth"] = width
    
    json_dat['shapes'] = shapes
    
    jsons.append(json_dat)
    idxs.append(imageName)
    #break
    
jsons = [jsons[i] for i in np.argsort(idxs)]
```

```python
write_to_jsonl(jsons, './work_dirs/convnext_large_exp_001/sub1.jsonl')
```
